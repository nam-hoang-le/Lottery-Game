# Import libraries

```python
import random
from IPython.display import clear_output
```

# Create important functions

### Create total money

```python
total_money = 1000000
```

### Shuffle prizes

```python
def shuffle_prizes():
    # FIRST PRIZE
    FIRST_PRIZE = random.randint(10,99)
    # SECOND PRIZE
    SECOND_PRIZE = random.randint(10,99)
    # THIRD PRIZE
    THIRD_PRIZE = random.randint(10,99)
    # FOURTH PRIZE
    FOURTH_PRIZE = random.randint(10,99)
    # FIFTH PRIZE
    FIFTH_PRIZE = random.randint(10,99)
    # SIXTH PRIZE
    SIXTH_PRIZE = random.randint(10,99)
    # SEVENTH PRIZE
    SEVENTH_PRIZE = random.randint(10,99)
    return FIRST_PRIZE, SECOND_PRIZE, THIRD_PRIZE, FOURTH_PRIZE, FIFTH_PRIZE, SIXTH_PRIZE, SEVENTH_PRIZE
```

### Print list of prizes

```python
def print_list_of_prize(list_of_prizes):
    print('-' * 20)
    print("---LIST OF PRIZES---")
    numb_prize = 1
    for numb_winning in list_of_prizes:
        print(f"Giai {numb_prize}: {numb_winning}")
        numb_prize += 1
    print('-' * 20)
```

### Enter lucky number and amount of money to play

Lucky number

```python
def enter_number_playing():
    while True:
        number_playing = input("Enter the lucky number you want to play (10-99): ")
        if number_playing.isdigit() and 10 <= int(number_playing) <= 99:
            return int(number_playing)
        else:
            print("Wrong format, please enter again!")
```

Money

```python
def enter_money_playing():
    while True:
        money_playing = input("Enter an amount of money you want to play (playing money < your total money): ")
        if money_playing.isdigit() and int(money_playing) < total_money:
            return int(money_playing)
        else:
            print("Wrong format or not enough money to play. Please enter again!")
```

### Handle your shuffle game  

```python
def shuffle_game(number_playing, money_playing):
    global total_money

    list_of_prize = shuffle_prizes()
    winning_numb_count = 0
    print(f"Lucky number you played: {number_playing}")
    print_list_of_prize(list_of_prize)
    for prize in list_of_prize:
        if number_playing == prize:
            winning_numb_count += 1
    if winning_numb_count > 0:
        total_money += (money_playing * winning_numb_count) * 70
        print(f"You have won {winning_numb_count} times!")
        print(f"Number of money you won {(money_playing * winning_numb_count) * 70}")
        print(f"Now you have {total_money}")
    else:
        print("You have lost")
        total_money -= money_playing
        print(f"Now you have {total_money}")
```

### Input more money

```python
def input_money():
    global total_money
    while True:
        money_in = int(input("Nhập số tiền mà bạn muốn nạp: "))
        total_money += money_in
        print(f"Now you have {total_money}")
        continue_input = int(input("Do you want to load more money? (0: No, 1: Yes)"))
        if continue_input == 0:
            break
        else:
            continue
```

### Menu

```python
def menu():
    print("LOTTERY GAME")
    print("1. Play game")
    print("2. Load money")
    print("3. Exit")
```

```python
def main():
    menu()
    while True:
        lua_chon = int(input("You choose: "))
        match lua_chon:
            case 1:
                clear_output()
                while True:
                    money_playing = enter_money_playing()
                    number_playing = enter_number_playing()
                    shuffle_game(money_playing, number_playing)
                    continue_play = int(input("Do you want to continue? (0: No, 1: Yes)"))
                    if continue_play == 0:
                        break
                clear_output(wait=True)
                main()
                break
            case 2:
                clear_output()
                input_money()
                clear_output(wait=True)
                main()
                break
            case 3:
                print("Bye, see you next time!")
                break
            case _:
                print("Wrong input, please enter again!")
```

```python
main()
```
